### Sorting and Ranking

12. Write a code snippet to group customers by segment and calculate the total spending amount for each segment, sorted in descending order of spending.

```python
spend.sort_values(by = 'Amount' , ascending = False ).head(10)
```

```python
# Merge spend with credit to get segment information
spend_with_segment = pd.merge(spend, credit[['Customer', 'Segment']], on='Customer', how='left')  # keep only spend records that have a matching customer in credit)

# Group by segment and calculate total spending
segment_spending = (spend_with_segment.groupby('Segment')['Amount'].sum().sort_values(ascending=False).reset_index())

segment_spending
```

### Data Merging and Joining
13. Merge the customer segment information from the `credit` dataframe with the total spending data to analyze spending trends across segments.
14. Merge the age group data with the spending data, then calculate and display the total spending amount for each age group.

```python
seg_spend = pd.merge(spent ,credit[['Customer' , 'Segment']], on = 'Customer')
seg_spend
```

```python
spend_with_age = pd.merge(spent, credit[['Customer', 'Age']], on='Customer', how='left')
bins = [0, 25, 35, 50, 100] 
labels = ['<25', '25-35', '36-50', '50+']
spend_with_age['Age_Group'] = pd.cut(spend_with_age['Age'], bins=bins, labels=labels, right=True)

age_group_spending = (spend_with_age.groupby('Age_Group' ,observed=True)['Amount'].sum().sort_values(ascending=False).reset_index())

age_group_spending
```
